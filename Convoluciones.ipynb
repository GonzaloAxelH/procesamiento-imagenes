```python
import matplotlib.pyplot as plt
import skimage.io as skio
import skimage.color as skcolor
import numpy as np
```

```python
iglesia=skio.imread("iglesia.jpg")
plt.imshow(iglesia)
```

```python
iglesia=skio.imread("iglesia.jpg")
plt.imshow(iglesia)
iglesiag=(skcolor.rgb2gray(iglesia)*255).astype(np.uint8)
plt.imshow(iglesiag,cmap="gray") 
print(iglesia.shape)
print(iglesiag)

```

```python
def filtrar(m,filtro):
  mr=np.multiply(m,filtro)
  return sum(sum(mr))

def convolucion(imagen,filtro):
  r=np.zeros(imagen.shape)
  fi,ci=imagen.shape
  for i in range(fi-2):
    for j in range(ci-2):
      r[i,j]=filtrar(imagen[i:i+3,j:j+3],filtro)
  return r
```

```python
fh=np.array([[-1,-2,-1],[0,0,0],[1,2,1]]) #horizontal
print(fh)   #como funcion la convolucion en matematicas 
```

```python
fh=np.array([[-1,-2,-1],[0,0,0],[1,2,1]]) #formas de convulcion
print(fh)                                 #buscar como funciona la convulcion matematicamente
resultado=convolucion(iglesiag,fh)
plt.imshow(resultado,cmap="gray")
```

```python
def fltrar(m,filtro):
  mr=np.multiply(m,filtro)
  return min(max(sum(sum(mr)),0),255) #es un max hasta 255 y nu minimo de 0 no se puede pasar de 255 ni menos de 0

def convolucion(imagen,filtro):
  r=np.zeros(imagen.shape)
  fi,ci=imagen.shape
  for i in range(fi-2):
    for j in range(ci-2):
      r[i,j]=filtrar(imagen[i:i+3,j:j+3],filtro)
  return r
```

```python
fh=np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
print(fh) #filtro horizontal
resultado=convolucion(iglesiag,fh)
plt.imshow(resultado,cmap="gray")
```
